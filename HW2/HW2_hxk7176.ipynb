```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset_url = 'https://raw.githubusercontent.com/m-mehdi/pandas_tutorials/main/weekly_stocks.csv'
df = pd.read_csv(dataset_url, parse_dates=['Date'], index_col='Date')
pd.set_option('display.max.columns', None) 
df.head()
```

```python
# 1. Line Plot: instead of MSFT plot for AAPL
df.plot(y='AAPL', figsize=(9,6))
```

```python
# 2. Multiline Plot: change the color of the three lines to blue, magenta and yellow.
df.plot.line(y=['FB', 'AAPL', 'MSFT'], figsize=(10,6), color=['blue', 'magenta', 'yellow'])
```

```python
# Name y-axis
df.plot(y='FB', figsize=(10,6), title='Facebook Stock', ylabel='USD')
```

```python
# three months Bar Plot
df_3Months = df.resample(rule='ME').mean()[-3:]
print(df_3Months)
```

```python
# Name y-axis
df_3Months.plot(kind='bar', figsize=(10,6), ylabel='Price')
```

```python
# horizontal bar chart
df_3Months.plot(kind='barh', figsize=(9,6))
```

```python
# stacked vertical
df_3Months.plot(kind='bar', stacked=True, figsize=(9,6))
```

```python
# Histogram
df[['MSFT', 'FB']].plot(kind='hist', bins=25, alpha=0.6, figsize=(9,6))
```

```python
# 3. Histogram: Plot histogram for all three stocks and change the interval to 15.
df[['MSFT', 'FB']].plot(kind='hist', bins=15, alpha=0.6, stacked=True, figsize=(9,6))
```

```python
# 4. Box Plot
df.plot(kind='box', figsize=(9,6))
```

```python
# 4. Box Plot
df.plot(kind='box', vert=False, figsize=(9,6))
```

```python
# Area Plot
df.plot(kind='area', figsize=(9,6))
```

```python
# Area Plot
df.plot(kind='area', stacked=False, figsize=(9,6))
```

```python
# Scatter Plot
# Scatter plots plot data points on the x and y axes to show the correlation between two variables. Like this:
df.plot(kind='scatter', x='MSFT', y='AAPL', figsize=(9,6), color='Green')
```

```python
# 5. Scatter Plot
# 5-1 'MSFT vs FB in blue color
df.plot(kind='scatter', x='MSFT', y='FB', figsize=(9,6), color='blue')
```

```python
# 5-2MSFT vs AAPL in red color
df.plot(kind='scatter', x='MSFT', y='AAPL', figsize=(9,6), color='red')
```
